```
import random
from matplotlib import pyplot as plt
```

```
def mean(arr):
  return sum(arr)/len(arr)
```

```
def variance(arr):
  mean_arr = mean(arr)
  return sum((element-mean_arr)**2 for element in arr)
```

```
def covariance(x, y):
  if len(x) != len(y):
    raise Exception('Both arrays must be of equal length')
  mean_x = mean(x)
  mean_y = mean(y)

  sum_values = 0
  for element_x, element_y in zip(x, y):
    sum_values += (element_x - mean_x)*(element_y - mean_y)
  return sum_values
```

```
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
```

```
data = []
for iteration in range(150):
  x = iteration
  y = x * 5 + random.randint(-50, 50)
  data.append(Point(x, y))
```

```
def linear_regression(arr):
  x = [element.x for element in arr]
  y = [element.y for element in arr]

  ## y = b0 + b1*x
  f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
  ax1.scatter(x, y)
  ax1.set_title('Plotting the points')

  mean_x = mean(x)
  mean_y = mean(y)

  b1 = covariance(x, y) / variance(x)
  b0 = mean_y - b1 * mean_x

  y_pred = [(b0 + b1 * element_x) for element_x in x]
  ax2.scatter(x, y)
  ax2.plot(x, y_pred, 'r')
  ax2.set_title('Plotting the points and fitting the line')
```

```
linear_regression(data)
```
